## 0. Setup — import libraries
Run this cell first. It loads the tools we need to create charts.

```python
# pandas  → handles data (like Excel, but in Python)
# matplotlib → draws charts
# seaborn → makes charts look nicer with less code

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns

# This makes all charts look clean and consistent
sns.set_theme(style='whitegrid', font_scale=1.1)
plt.rcParams['figure.dpi'] = 120
plt.rcParams['savefig.bbox'] = 'tight'

print('Libraries loaded successfully!')
```

---
## 1. Programming Language Trends
### 1a. Top 10 most used languages (current year)

```python
# --- DATA ---
# Each language and how many respondents said they use it
languages_current = pd.DataFrame({
    'Language': [
        'JavaScript', 'SQL', 'HTML/CSS', 'TypeScript', 'Python',
        'Bash/Shell', 'C#', 'Java', 'PHP', 'PowerShell'
    ],
    'Respondents': [
        14943, 12602, 12410, 10709, 9590,
        7244, 6340, 5982, 4644, 3438
    ]
})

# Sort from highest to lowest
languages_current = languages_current.sort_values('Respondents', ascending=False)

# --- CHART ---
fig, ax = plt.subplots(figsize=(10, 6))

bars = ax.bar(
    languages_current['Language'],
    languages_current['Respondents'],
    color=sns.color_palette('Blues_d', len(languages_current))
)

# Add the number on top of each bar
for bar in bars:
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() + 150,
        f'{int(bar.get_height()):,}',
        ha='center', va='bottom', fontsize=9
    )

ax.set_title('Top 10 Most Used Programming Languages (2024)', fontsize=14, pad=15)
ax.set_xlabel('Programming Language')
ax.set_ylabel('Number of Respondents')
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
plt.xticks(rotation=15, ha='right')
plt.tight_layout()

# Save the chart as an image file
plt.savefig('chart_01_languages_current.png')
plt.show()
print('Chart saved!')
```

### 1b. Top 10 languages to watch next year

```python
# --- DATA ---
languages_future = pd.DataFrame({
    'Language': [
        'JavaScript', 'SQL', 'TypeScript', 'HTML/CSS', 'Python',
        'Go', 'Rust', 'C#', 'Bash/Shell', 'Java'
    ],
    'Respondents': [
        11541, 10944, 10437, 10016, 8919,
        5661, 5597, 5590, 5582, 4048
    ]
})

languages_future = languages_future.sort_values('Respondents', ascending=False)

# --- CHART ---
fig, ax = plt.subplots(figsize=(10, 6))

bars = ax.bar(
    languages_future['Language'],
    languages_future['Respondents'],
    color=sns.color_palette('Purples_d', len(languages_future))
)

for bar in bars:
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() + 100,
        f'{int(bar.get_height()):,}',
        ha='center', va='bottom', fontsize=9
    )

ax.set_title('Top 10 Programming Languages to Watch in 2025', fontsize=14, pad=15)
ax.set_xlabel('Programming Language')
ax.set_ylabel('Number of Respondents')
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
plt.xticks(rotation=15, ha='right')
plt.tight_layout()

plt.savefig('chart_02_languages_future.png')
plt.show()
print('Chart saved!')
```

## 2. Database Trends
### 2a. Top 10 most used databases (current year)

```python
# --- DATA ---
databases_current = pd.DataFrame({
    'Database': [
        'PostgreSQL', 'MySQL', 'SQLite', 'MongoDB', 'Microsoft SQL Server',
        'Redis', 'MariaDB', 'Elasticsearch', 'DynamoDB', 'Oracle'
    ],
    'Respondents': [
        11514, 8556, 7021, 5930, 5870,
        5814, 3994, 3491, 2268, 1907
    ]
})

# Sort ascending so the longest bar is at the top (horizontal chart)
databases_current = databases_current.sort_values('Respondents', ascending=True)

# --- CHART ---
# Horizontal bar chart — easier to read database names
fig, ax = plt.subplots(figsize=(10, 6))

bars = ax.barh(
    databases_current['Database'],
    databases_current['Respondents'],
    color=sns.color_palette('Greens_d', len(databases_current))
)

# Add number labels at the end of each bar
for bar in bars:
    ax.text(
        bar.get_width() + 80,
        bar.get_y() + bar.get_height() / 2,
        f'{int(bar.get_width()):,}',
        va='center', fontsize=9
    )

ax.set_title('Top 10 Most Used Databases (2024)', fontsize=14, pad=15)
ax.set_xlabel('Number of Respondents')
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
plt.tight_layout()

plt.savefig('chart_03_databases_current.png')
plt.show()
print('Chart saved!')
```

### 2b. Top 10 most desired databases (next year)

```python
# --- DATA ---
databases_future = pd.DataFrame({
    'Database': [
        'PostgreSQL', 'Redis', 'SQLite', 'MySQL', 'MongoDB',
        'Microsoft SQL Server', 'Elasticsearch', 'MariaDB', 'DynamoDB', 'Supabase'
    ],
    'Respondents': [
        12193, 6384, 6295, 6204, 5618,
        4345, 3665, 3078, 2154, 1623
    ]
})

databases_future = databases_future.sort_values('Respondents', ascending=True)

# --- CHART ---
fig, ax = plt.subplots(figsize=(10, 6))

bars = ax.barh(
    databases_future['Database'],
    databases_future['Respondents'],
    color=sns.color_palette('Oranges_d', len(databases_future))
)

for bar in bars:
    ax.text(
        bar.get_width() + 80,
        bar.get_y() + bar.get_height() / 2,
        f'{int(bar.get_width()):,}',
        va='center', fontsize=9
    )

ax.set_title('Top 10 Most Desired Databases for 2025', fontsize=14, pad=15)
ax.set_xlabel('Number of Respondents')
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
plt.tight_layout()

plt.savefig('chart_04_databases_future.png')
plt.show()
print('Chart saved!')
```

## 3. Salary by Programming Language

```python
# --- DATA ---
salary_data = pd.DataFrame({
    'Language': [
        'Swift', 'Python', 'C++', 'JavaScript', 'Java',
        'Go', 'R', 'C#', 'SQL', 'PHP'
    ],
    'Avg_Salary_USD': [
        130801, 114383, 113883, 110961, 101013,
        94082, 92037, 88726, 84793, 84727
    ]
})

salary_data = salary_data.sort_values('Avg_Salary_USD', ascending=False)

# --- CHART ---
fig, ax = plt.subplots(figsize=(11, 6))

# Colour the top 3 differently to highlight them
colors = ['#2E86AB' if i < 3 else '#A8DADC' for i in range(len(salary_data))]

bars = ax.bar(
    salary_data['Language'],
    salary_data['Avg_Salary_USD'],
    color=colors
)

# Add salary labels on top of each bar
for bar in bars:
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() + 600,
        f'${int(bar.get_height()):,}',
        ha='center', va='bottom', fontsize=8.5
    )

ax.set_title('Average Annual Salary by Programming Language (USD)', fontsize=14, pad=15)
ax.set_xlabel('Programming Language')
ax.set_ylabel('Average Annual Salary (USD)')
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${int(x):,}'))
ax.set_ylim(0, 145000)

# Add a legend explaining the colour difference
from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor='#2E86AB', label='Top 3 highest paying'),
    Patch(facecolor='#A8DADC', label='Others')
]
ax.legend(handles=legend_elements, loc='upper right')

plt.tight_layout()
plt.savefig('chart_05_salary_by_language.png')
plt.show()
print('Chart saved!')
```

## 4. Demographics
### 4a. Respondents by age group

```python
# --- DATA ---
age_data = pd.DataFrame({
    'Age Group': [
        '25–34 years', '35–44 years', '18–24 years',
        '45–54 years', 'Other / prefer not to say'
    ],
    'Percentage': [41.3, 27.3, 15.9, 10.9, 4.6]
})

# --- CHART ---
fig, ax = plt.subplots(figsize=(8, 8))

colors = ['#264653', '#2A9D8F', '#E9C46A', '#F4A261', '#E76F51']

wedges, texts, autotexts = ax.pie(
    age_data['Percentage'],
    labels=age_data['Age Group'],
    autopct='%1.1f%%',       # shows percentage on each slice
    startangle=140,
    colors=colors,
    pctdistance=0.75,
    wedgeprops=dict(edgecolor='white', linewidth=2)
)

# Make percentage text bold and readable
for autotext in autotexts:
    autotext.set_fontsize(10)
    autotext.set_fontweight('bold')
    autotext.set_color('white')

ax.set_title('Respondent Distribution by Age Group', fontsize=14, pad=20)
plt.tight_layout()

plt.savefig('chart_06_age_distribution.png')
plt.show()
print('Chart saved!')
```

### 4b. Respondents by education level

```python
# --- DATA ---
education_data = pd.DataFrame({
    'Education': [
        "Bachelor's degree",
        "Master's degree",
        'Some college',
        'Secondary school',
        'Professional degree',
        'Associate degree',
        'Something else',
        'Primary school'
    ],
    'Respondents': [8629, 5060, 2476, 1143, 661, 634, 159, 112]
})

education_data = education_data.sort_values('Respondents', ascending=True)

# --- CHART ---
fig, ax = plt.subplots(figsize=(10, 6))

bars = ax.barh(
    education_data['Education'],
    education_data['Respondents'],
    color=sns.color_palette('mako_r', len(education_data))
)

for bar in bars:
    ax.text(
        bar.get_width() + 40,
        bar.get_y() + bar.get_height() / 2,
        f'{int(bar.get_width()):,}',
        va='center', fontsize=9
    )

ax.set_title('Respondent Distribution by Education Level', fontsize=14, pad=15)
ax.set_xlabel('Number of Respondents')
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
plt.tight_layout()

plt.savefig('chart_07_education.png')
plt.show()
print('Chart saved!')
```
